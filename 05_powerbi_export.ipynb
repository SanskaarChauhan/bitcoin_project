```python
# Create Power BI optimized exports
import pandas as pd

# Load your cleaned data with proper dtype handling
sentiment_df = pd.read_csv('cleaned_sentiment.csv')
merged_df = pd.read_csv('cleaned_merged_trades.csv', low_memory=False)
daily_stats = pd.read_csv('daily_trader_stats.csv')
trader_stats = pd.read_csv('trader_performance_summary.csv')

# Export for Power BI
sentiment_df.to_csv('powerbi_sentiment.csv', index=False)
merged_df.to_csv('powerbi_trades.csv', index=False)
daily_stats.to_csv('powerbi_daily_stats.csv', index=False)
trader_stats.to_csv('powerbi_trader_stats.csv', index=False)

print("Power BI files exported successfully")
print(f"Files created:")
print(f"- powerbi_sentiment.csv ({len(sentiment_df)} rows)")
print(f"- powerbi_trades.csv ({len(merged_df)} rows)")
print(f"- powerbi_daily_stats.csv ({len(daily_stats)} rows)")
print(f"- powerbi_trader_stats.csv ({len(trader_stats)} rows)")
```
